```python
fitsfile = './Einstein.fits'
map_tree = '../res_file/map_20240731_WCDA.root'
det_res = '../res_file/res_20240731_WCDA.root'
mapname = './data/Einstein.hd5'

from scipy.special import k0

def diffusion(cx, cy, theta, e, width ,pw, xsize):
    '''
    cx, cy: center, float, 0-1
    '''
    src = np.zeros((xsize, xsize))
    x = np.arange(xsize)*pw
    ix = np.arange(xsize)
    xx,yy=np.meshgrid(x,x)
    cpx = np.interp(cx, ix/xsize, x)+ np.pi* 1e-2
    cpy = np.interp(cy, ix/xsize, x)+ np.pi* 1e-2
    rr = np.sqrt(((xx-cpx) * np.cos(theta) + (yy-cpy) * np.sin(theta))**2 +
                 e* (-(xx-cpx) * np.sin(theta) + (yy-cpy) * np.cos(theta))**2)
    src = np.exp( -rr/width)* k0(rr)
    srcsum = src.sum()
    assert np.isfinite(srcsum), "Sum must be finite"
    
    return src/srcsum * (pw/180 * np.pi)**2

data = diffusion(np.random.uniform(0.5,0.52),np.random.uniform(0.5,0.52),0,1,2.2,0.1,150)

print(data.shape) # random map

xsize = 64  # size of final blurred array 
# Create fits file with 2darray
data2fits(data, 150, 10, fitsfile, xsize_fin = xsize, pixelw = 0.1)
# Now the fits file is saved. 


```

```python
# Perform detres simulation
%matplotlib inline
img = respconv(
    fitsfile, map_tree, det_res, mapname,
    xsize, srcname='src',
    plindex=2.63, piv=1e9, specK=2e-25,
    bins='0'
        )
```
