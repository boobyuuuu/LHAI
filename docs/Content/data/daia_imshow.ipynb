```python
import numpy as np
import matplotlib.pyplot as plt

# 
data = np.load('poisson_src_bkg.pkl.npy')  # shape: (2, 10000, 64, 64)

# 10
imgs1 = data[:10, 0]
imgs2 = data[:10, 1]

#  colorbar
fig, axes = plt.subplots(10, 2, figsize=(6, 12))  # 210

for i in range(10):
    # 
    im1 = axes[i, 0].imshow(imgs1[i], cmap='viridis')
    axes[i, 0].axis('off')
    plt.colorbar(im1, ax=axes[i, 0], fraction=0.046, pad=0.04)

    # 
    im2 = axes[i, 1].imshow(imgs2[i], cmap='viridis')
    axes[i, 1].axis('off')
    plt.colorbar(im2, ax=axes[i, 1], fraction=0.046, pad=0.04)

plt.tight_layout()
# plt.savefig('imshow',dpi = 300)
```
